## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Convert 'normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', and 'price' to numeric, coercing errors to NaN
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'], errors='coerce')
df['bore'] = pd.to_numeric(df['bore'], errors='coerce')
df['stroke'] = pd.to_numeric(df['stroke'], errors='coerce')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['peak-rpm'] = pd.to_numeric(df['peak-rpm'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Handle 'num-of-doors' separately first: map string representations to numbers, then convert to numeric
df['num-of-doors'] = df['num-of-doors'].replace({'two': 2, 'four': 4})
df['num-of-doors'] = pd.to_numeric(df['num-of-doors'], errors='coerce')

# Handle missing values:
# For 'normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm': impute with mean
for col in ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm']:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mean()) # Assign result back to avoid FutureWarning

# For 'num-of-doors', fill with the mode (most frequent value)
if df['num-of-doors'].isnull().any():
    mode_val_doors = df['num-of-doors'].mode()
    if not mode_val_doors.empty:
        df['num-of-doors'] = df['num-of-doors'].fillna(mode_val_doors[0])

# Drop rows where 'price' is NaN, as price is the target variable and imputation might be misleading
df.dropna(subset=['price'], inplace=True)

# Verify data types and nulls after wrangling
print("Data types after wrangling:")
df.info()
print("\nMissing values after wrangling:")
print(df.isnull().sum())
````
